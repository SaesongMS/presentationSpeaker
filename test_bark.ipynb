```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
```

```python
# download and load all models
preload_models()
```

```python
#pl speakers, that are quite good for me: v2/pl_speaker_0, v2/pl_speaker_3

# generate audio from text
text_prompt = """
    Jestem bark, podaj mi tekst, a ja go przeczytam.
    Zrobię to najlepiej jak potrafię, ale pamiętaj, że jestem tylko programem komputerowym.
"""
audio_array = generate_audio(text_prompt, history_prompt="v2/pl_speaker_3")
# audio_array = generate_audio(text_prompt) # use default speaker
  
# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
# save audio to disk
write_wav("generations/bark_generation3.mp3", SAMPLE_RATE, audio_array)
```
